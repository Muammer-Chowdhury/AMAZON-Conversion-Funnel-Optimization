```python
# --- Conversion Funnel Analysis (Amazon-style) ---

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

stages = ["Visit", "Add to Cart", "Checkout", "Purchase"]
n_users = 1000

# Simulate funnel data
data = []
for user in range(n_users):
    prob = 1
    for stage in stages:
        if np.random.rand() < prob:
            data.append([user, stage])
        prob *= np.random.uniform(0.6, 0.9)  # simulate drop-off

df = pd.DataFrame(data, columns=["user_id", "stage"])

# Count users per stage
funnel = df["stage"].value_counts().reindex(stages, fill_value=0)

conversion_rates = funnel / funnel.iloc[0] * 100
print(conversion_rates)

# Visualization
plt.figure(figsize=(7,5))
plt.plot(stages, conversion_rates, marker="o", color="green")
plt.title("Conversion Funnel (Simulated E-commerce)")
plt.xlabel("Funnel Stage")
plt.ylabel("Conversion Rate (%)")
plt.ylim(0, 100)
plt.show()
```
